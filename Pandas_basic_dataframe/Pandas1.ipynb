## Position-based Indexing and Slicing with iloc[]

```python
summer = pd.read_csv("summer.csv")
```

```python
summer
```

```python
summer.info()
```

```python
summer = pd.read_csv("summer.csv", index_col = "Athlete")   # "Athlete"을 기준으로 DataFrame 구성
```

```python
summer
```

```python
summer.info()
```

### Selecting Rows with iloc[]

```python
summer.iloc[0]     # index 0행을 가져옴
```

```python
type(summer.iloc[0])
```

```python
summer.iloc[1]       # index 1행을 가져옴
```

```python
summer.iloc[-1]       # index -1행을 가져옴 - 맨 마지막 행
```

```python
summer.iloc[[1,2,3]]     # index 1,2,3행을 가져옴 - DataFrame 형태
```

```python
summer.iloc[1:3]         # index 1부터 3-1=2행까지를 가져옴 - DataFrame 형태
```

```python
summer.iloc[1:6]         # index 1부터 5행까지를 가져옴 - DataFrame 형태
```

```python
summer.iloc[-5:]         # index -5부터 맨 마지막 행까지를 가져옴
```

```python
summer.iloc[-5:-1]       # index -5부터 맨 마지막 바로 앞 행까지를 가져옴
```

```python
summer.iloc[[2,45, 31100]]    # index 2행, 45행, 31100행을 가져옴 - 특정 행의 추출에는 이 방식이 더 유리
```

```python
summer.iloc[0, 4]       # 0행의 index 4에 해당하는 열의 정보
```

```python
summer.iloc[0, :3]      # 0행의 0~3 index 열의 정보
```

```python
summer.iloc[0, [0,2,5,7]]    # 0행의 0,2,5,7 index 열의 정보
```

```python
summer.iloc[34:39, [0,2,5,7]]   # index 34~38행의 0,2,5,7 index 열의 정보
```

```python
summer.iloc[:, 4]             # 모든 행의 index 4에 해당하는 정보
```

```python
summer.iloc[:, 4].equals(summer.Country)
```

```python
summer["Country"]
```

```python
summer.Country
```

## Label-based Indexing and Slicing with loc[]

```python
summer
```

### Selecting Rows with loc[]

```python
summer.iloc[2]
```

```python
summer.loc["DRIVAS, Dimitrios"]      # Label-based Indexing
```

```python
summer.loc["PHELPS, Michael"]      # Label-based Indexing
```

### Indexing/Slicing Rows and Columns with loc[]

```python
summer.loc["PHELPS, Michael","Medal"]   # PHELPS가 받은 Medal의 모든게 등장
```

```python
summer.loc["PHELPS, Michael",["Medal", "Event"]]
```

```python
summer.loc[["PHELPS, Michael", "LEWIS, Carl"],["Medal", "Event"]]
```

```python
summer.loc[:,["Medal", "Event"]]      # 모든 행에 대한 Medal, Event의 정보
```

```python
summer.loc[:"CHASAPIS, Spiridon"]     # 처음부터 Spiridon까지의 모든 정보
# iloc은 마지막이 제외, loc은 마지막까지 포함
```

```python
summer.loc[:"PHELPS, Michael"]  
# PHELPS, Michael과 같이 중복되는 경우는 어디까지를 가져와야할지 모르기 때문에 오류메세지 출력
```

```python
summer.head(20)
```

```python
summer.loc["DRIVAS, Dimitrios":"BLAKE, Arthur"]     # loc이므로 양쪽 다 포함
```

```python
summer.loc["DRIVAS, Dimitrios":"BLAKE, Arthur", "Year":"Discipline"]  
# 행과 열을 한 번에 Slicing
```

### Indexing and Slicing with reindex()

```python
summer = pd.read_csv("summer.csv")
```

```python
summer
```

```python
summer.loc[[0,5,30000], ["Athlete","Medal"]]
```

```python
summer.reindex(index = [0,5,30000,40000], columns = ["Athlete","Medal"])
# 기존에 있지 않은 index를 만들어서 넣는 방법
```
