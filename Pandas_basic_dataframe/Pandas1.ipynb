## Make it easy: TAB Completion and Tooltip

```python
import pandas as pd
```

```python
pd.read_csv("titanic.csv")    ## TAB키를 활용한 내장함수 불러오기
```

```python
titanic.sort_values(by="age", ascending=False)
#ascending=False - 내림차순(default값 : True)
##shift + TAB, shift + TAB + TAB키 누르면 사용할만한 문법(Tooltip) 나옴
```
