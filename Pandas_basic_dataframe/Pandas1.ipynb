## Position-based Indexing and Slicing with iloc[]

```python
summer = pd.read_csv("summer.csv")
```

```python
summer
```

```python
summer.info()
```

```python
summer = pd.read_csv("summer.csv", index_col = "Athlete")   # "Athlete"을 기준으로 DataFrame 구성
```

```python
summer
```

```python
summer.info()
```

### Selecting Rows with iloc[]

```python
summer.iloc[0]     # index 0행을 가져옴
```

```python
type(summer.iloc[0])
```

```python
summer.iloc[1]       # index 1행을 가져옴
```

```python
summer.iloc[-1]       # index -1행을 가져옴 - 맨 마지막 행
```

```python
summer.iloc[[1,2,3]]     # index 1,2,3행을 가져옴 - DataFrame 형태
```

```python
summer.iloc[1:3]         # index 1부터 3-1=2행까지를 가져옴 - DataFrame 형태
```

```python
summer.iloc[1:6]         # index 1부터 5행까지를 가져옴 - DataFrame 형태
```

```python
summer.iloc[-5:]         # index -5부터 맨 마지막 행까지를 가져옴
```

```python
summer.iloc[-5:-1]       # index -5부터 맨 마지막 바로 앞 행까지를 가져옴
```

```python
summer.iloc[[2,45, 31100]]    # index 2행, 45행, 31100행을 가져옴 - 특정 행의 추출에는 이 방식이 더 유리
```
